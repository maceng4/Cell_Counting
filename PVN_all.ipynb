```python
from math import sqrt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray
import numpy as np

import matplotlib.pyplot as plt
```

```python
# PVN IMAGE
from PIL import Image
import numpy
img_pvn = Image.open("PVN.png")

# PVN IMAGE RGBA -> RGB
img_pvn.load()
background = Image.new("RGB", img_pvn.size, (255, 255, 255))
background.paste(img_pvn, mask = img_pvn.split()[3])
background.save("img_pvn_final.jpg", "JPEG", quality=100)

rgb_image = Image.open("img_pvn_final.jpg")

# PVN IMAGE
np_img_pvn = numpy.array(rgb_image)
image_gray_pvn = rgb2gray(np_img_pvn)
```

```python
# Load Sample images
image = data.hubble_deep_field()[0:500, 0:500]
image_gray = rgb2gray(image)

blobs_log = blob_log(image_gray, max_sigma=30, num_sigma=10, threshold=.1)

# Compute radii in the 3rd column.
blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

blobs_dog = blob_dog(image_gray, max_sigma=30, threshold=.1)
blobs_dog[:, 2] = blobs_dog[:, 2] * sqrt(2)

blobs_doh = blob_doh(image_gray, max_sigma=30, threshold=.01)
```

```python
#SAMPLE IMAGE

# Compute radii in the 3rd column.
#LOG
blobs_log_pvn = blob_log(image_gray_pvn, max_sigma=30, num_sigma=10, threshold=.1)
blobs_log_pvn[:, 2] = blobs_log_pvn[:, 2] * sqrt(2)

#DOG
blobs_dog_pvn = blob_dog(image_gray_pvn, max_sigma=30, threshold=.1)
blobs_dog_pvn[:, 2] = blobs_dog_pvn[:, 2] * sqrt(2)

#DOH
blobs_doh_pvn = blob_doh(image_gray_pvn, max_sigma=30, threshold=.01)

blobs_list = [blobs_log, blobs_log_pvn, blobs_dog, blobs_dog_pvn, blobs_doh, blobs_doh_pvn]
colors = ['yellow', 'yellow', 'lime', 'lime', 'red', 'red']
titles = ['Laplacian of Gaussian', 'PVN LOG Image', 'Difference of Gaussian',
          'PVN DOG Image', 'Determinant of Hessian', 'PVN DOH Image']
sequence = zip(blobs_list, colors, titles)

fig, axes = plt.subplots(1, 6, figsize=(9, 3), sharex=True, sharey=True)
ax = axes.ravel()

for idx, (blobs, color, title) in enumerate(sequence):
    ax[idx].set_title(title)
    ax[idx].imshow(image)
    for blob in blobs:
        y, x, r = blob
        c = plt.Circle((x, y), r, color=color, linewidth=2, fill=False)
        ax[idx].add_patch(c)
    ax[idx].set_axis_off()

plt.tight_layout()
plt.show()
```
